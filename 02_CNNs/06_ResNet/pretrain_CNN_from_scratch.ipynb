# 3. Dataset

```python
stats = ((0.5071, 0.4867, 0.4408), (0.2675, 0.2565, 0.2761))

train_transform = transforms.Compose([
    transforms.AutoAugment(transforms.AutoAugmentPolicy.CIFAR10),
    transforms.RandomHorizontalFlip(),
    transforms.ToTensor(),
    transforms.Normalize(*stats)
])

test_transform = transforms.Compose([
    transforms.ToTensor(),
    transforms.Normalize(*stats)
])
```

```python
import copy

# Download the full train dataset
full_train_dataset = datasets.CIFAR100(root=DATA_PATH, train=True, download=True, transform=train_transform)

# Split the full train dataset into train and validation subset
train_size = int(0.8 * len(full_train_dataset))
val_size = len(full_train_dataset) - train_size

train_subset, val_subset = random_split(full_train_dataset, (train_size, val_size))

# Change the transform of val_subset to test_transform
val_subset.dataset = copy.copy(full_train_dataset)
val_subset.dataset.transform = test_transform

# Download the test dataset
test_dataset = datasets.CIFAR100(root=DATA_PATH, train=False, download=True, transform=test_transform)
```

```python
train_loader = DataLoader(train_subset, batch_size=BATCH_SIZE, shuffle=True,
                        num_workers=4, pin_memory=True)

val_loader = DataLoader(val_subset, batch_size=BATCH_SIZE, shuffle=True,
                        num_workers=4, pin_memory=True)

test_loader = DataLoader(test_dataset, batch_size=BATCH_SIZE, shuffle=True,
                        num_workers=4, pin_memory=True)
```
